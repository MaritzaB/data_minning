# Generación del cubo de datos de los datos de trayectorias de aves marinas

## Introducción

La creación de un cubo de datos, se hace con la finalidad de tener diferentes
vistas con los valores agregados de los datos, para poder hacer un análisis más
rápido y eficiente de los datos.

Para el caso de la base de datos de trayectorias de aves marinas, se creará un
cubo de datos con las siguientes características:

* **Dimensión de tiempo**: Se creará una dimensión de tiempo con los años.
* **Dimensión de especie**: Se creará una dimensión de especie con las especies
  de aves marinas.
* **Dimensión de temporada reproductiva**: Se creará una dimensión de temporada
  reproductiva con las temporadas reproductivas de las especies de aves marinas.

* **Medida del área de la envolvente convexa al grupo de puntos de la dimensión**:
  La envolvente convexa es una figura geométrica que contiene a un conjunto de
  puntos, de tal manera que cada punto del conjunto está en el interior de la
  figura o en su frontera. En este caso, se calculará el área de la envolvente
  convexa de los puntos de la dimensión de tiempo, de la dimensión de especie y
  de la dimensión de temporada reproductiva.

# Consultas SQL para la creación del cubo de datos

```sql
-- Cuboide base 3D
create view view_3D_nys as
select
	name as albatros_id,
	substring(date, 1, 4)  as año,
	season as temporada,
	st_convexhull(st_collect(geom)),
	st_area(st_convexhull(st_collect(geom))::geography)/10000 as ConvexHullArea_ha2
from albatros_seasons as2
group by albatros_id, año, temporada

-- Cuboides 2D
-- ID, año, all temporadas
create view view_2D_110 as
select 
	name as albatros_id,
	substring(date, 1, 4) as año,
	'ALL' as temporada,
	st_convexhull(st_collect(geom)),
	st_area(
	st_convexhull(st_collect(geom))::geography)/10000 as ConvexHullArea_ha2
from albatros_seasons as2 
group by albatros_id, año
order by ConvexHullArea_ha2 desc;

-- ID, all año,temporadas
create view view_2D_101 as
select 
	name as albatros_id,
	'ALL' as año,
	season as temporada,
	st_convexhull(st_collect(geom)),
	st_area(st_convexhull(st_collect(geom))::geography)/10000 as ConvexHullArea_ha2
from albatros_seasons as2 
group by albatros_id, temporada;


-- all ID, año, temporadas
create view view_2D_011 as
select 
	'ALL' as albatros_id,
	substring(date, 1,4) as año,
	season as temporada,
	st_convexhull(st_collect(geom)),
	st_area(st_convexhull(st_collect(geom))::geography)/10000 as ConvexHullArea_ha2
from albatros_seasons as2 
group by año, season;

-- Cuboide 1D
create view view_1D_001 as
select 
	'ALL' as id,
	'ALL' as año,
	season as temporada,
	st_convexhull(st_collect(geom)),
	st_area(st_convexhull(st_collect(geom))::geography)/10000 as ConvexHullArea_ha2
from albatros_seasons as2 
group by temporada;

create view view_1D_010 as
select 
	'ALL' as id,
	substring(date, 1, 4)  as año,
	'ALL' as temporada,
	st_convexhull(st_collect(geom)),
	st_area(st_convexhull(st_collect(geom))::geography)/10000 as ConvexHullArea_ha2
from albatros_seasons as2 
group by año;

create view view_1D_100 as
select 
	name as id,
	'ALL' as año,
	'ALL' as temporada,
	st_convexhull(st_collect(geom)),
	st_area(st_convexhull(st_collect(geom))::geography)/10000 as ConvexHullArea_ha2
from albatros_seasons as2 
group by name;
```
