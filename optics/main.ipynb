```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons, make_circles
import warnings
warnings.filterwarnings("ignore") 
```

```python
# generate 2d classification dataset
X, y = make_blobs(n_samples=30, centers=4, n_features=2, cluster_std=1.8, random_state=42)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title('Blobs with 4 clusters')
plt.savefig('varied.png')
```

```python

from optics import OPTICS, plot_clust

eps_radio = 2
ClustDist, CoreDist = OPTICS(X, eps=eps_radio, minPTS=3, plot=True, plot_reach=True)
```

```python
plot_clust(X, ClustDist, CoreDist, eps=eps_radio, eps_db=1.9)
```
